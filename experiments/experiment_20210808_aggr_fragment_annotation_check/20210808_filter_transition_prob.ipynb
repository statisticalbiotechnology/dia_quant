### We can see that the spectral library built using dia-umpire has different count of transitions per precursor, and that the original library search (with LFQBench database has 6 transition per precursor). This causes msStats to get memory error. We need to filter.

```python
x = df.aggr_Fragment_Annotation[0]
```

```python
x
```

```python
def transition_count(x):
    return len(x.split(";"))
```

```python
df["transition_count"] = df.aggr_Fragment_Annotation.map(transition_count)
```
