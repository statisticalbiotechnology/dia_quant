# m_score
According to [SWATH2stats vignette](https://www.bioconductor.org/packages/release/bioc/vignettes/SWATH2stats/inst/doc/SWATH2stats_vignette.pdf) the m_score corresponds to "an m-score threshold of e.g.  0.01 (corresponding to 1 % model FDR)." It seems to be on peptide-level.

```python
df_osw_filtered = df_osw[df_osw["m_score"] < 0.05]
```

```python
df_osw_filtered
```

It seems to be pre-filtered by OpenSwathWorkflow at 5% FDR at peptide-level.

```python
df_osw_filtered = df_osw[df_osw["m_score"] < 0.01]
```

```python
df_osw_filtered
```

```python
df_osw_filtered["species"] = df_osw_filtered.ProteinName.map(get_species)
```

```python
def get_run(x):
    return x.split("/")[1].split("_")[0].split("e")[1]
```

```python
df_osw_filtered["run"] = df_osw_filtered.filename.map(get_run)
```

```python
def get_condition(x):
    #x = int(x)
    A = [str(1),str(2),str(3),str(4)]
    B = [str(5),str(6),str(7),str(8)]
    if x in A:
        return "A"
    elif x in B:
        return "B"
    else:
        return np.nan
```

```python
df_osw_filtered["condition"] = df_osw_filtered.run.map(get_condition)
```

# OSW analysis

There could be three different imputations:
- NaN - remove proteins
- 0 - assume that NaN are because these are very low intensities
- mean - just take the average of all

# OSW - NaN remove proteins analysis

```python
prot = df_osw_analysis.protein.unique()[0]
```

```python
prot_df = df_osw_analysis[df_osw_analysis.protein == prot]
```

```python
prot_df["a"]
```

```python
a = prot_df[prot_df.condition == "A"].sum_of_peptide_intensities.mean()
```

```python
b = prot_df[prot_df.condition == "B"].sum_of_peptide_intensities.mean()
```

```python
def log2fc(a,b):
    return np.log2(b) - np.log2(a)
```

```python
prot_array = []
log_fold_change_array = []

for prot in df_osw_analysis.protein.unique():
    prot_df = df_osw_analysis[df_osw_analysis.protein == prot]
    a = prot_df[prot_df.condition == "A"].sum_of_peptide_intensities.mean()
    b = prot_df[prot_df.condition == "B"].sum_of_peptide_intensities.mean()
    log_fold_change = log2fc(a,b)
    prot_array.append(prot)
    log_fold_change_array.append(log_fold_change)

    
```

```python
df_osw_de = pd.DataFrame(np.array([prot_array, log_fold_change_array]).T, columns = ["protein", "log2fc"])
```

```python
df_osw_de["species"] = df_osw_de.protein.map(get_species)
```

```python
df_osw_de = df_osw_de.replace('nan', np.nan)
df_osw_de = df_osw_de.dropna() #remove nan
```

```python
df_osw_de.log2fc = df_osw_de.log2fc.astype(float)
```

```python
diff_exp_treshold = 1.0
```

```python
(df_osw_de.log2fc > diff_exp_treshold).sum()
```

```python
print("OSW diff. exp. prot at 1% FDR (m-score) and 1.0 log2fc treshold" + " : " + str((df_osw_de.log2fc > diff_exp_treshold).sum()))
```
